```python
import pandas as pd
```

```python
df_meta = pd.read_excel('/mnt/f/OneDrive/文档（科研）/P0_幽门螺旋杆菌/基础信息表/HP数据收集2.xlsx', sheet_name='HP数据收集')
```

```python
df_meta = df_meta.loc[:,['ID','Chromopainter4','Country','Continent']]
df_meta['ID'] = df_meta['ID'].astype(str)
```

```python
df_pca = pd.read_csv('/mnt/d/幽门螺旋杆菌/Script/分析结果/3-PCA/7544_整体_maf99/hpglobal_LD_PCA.eigenvec',sep=' ',header=None).drop(columns=0)
# 将第一列设置为ID
df_pca.columns = ['ID'] + [f'PC{i+1}' for i in range(df_pca.shape[1] - 1)]
df_pca['ID'] = df_pca['ID'].astype(str)
df_pca
```

```python
# 合并df_meta和df_pca
df_pca = df_pca.merge(df_meta, on='ID', how='left')
df_pca
```

```python
df_pca['Class_small'] = df_pca['Continent'] + '_' + df_pca['Country']
df_pca
```

```python
df_pca.rename(columns={'Continent': 'Class_big'}, inplace=True)
df_pca
```

```python
df_pca.to_csv('../data/example_pca.csv',index=False)
```
